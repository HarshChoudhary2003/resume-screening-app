```python
import re
import string

def cleanResume(txt):
    # Remove links
    cleanTxt = re.sub(r'http\S+\s', ' ', txt)
    
    # Remove RT (retweets) or standalone 'S+'
    cleanTxt = re.sub(r'RT|S+', ' ', cleanTxt)
    
    # Remove @mentions
    cleanTxt = re.sub(r'@\S+', ' ', cleanTxt)
    
    # Remove hashtags
    cleanTxt = re.sub(r'#\S+\s', ' ', cleanTxt)
    
    # Remove punctuation
    cleanTxt = re.sub(r'[%s]' % re.escape(string.punctuation), ' ', cleanTxt)
    
    # Remove non-ASCII characters (like emojis)
    cleanTxt = re.sub(r'[^\x00-\x7f]', ' ', cleanTxt)
    
    # Replace multiple spaces with a single space
    cleanTxt = re.sub(r'\s+', ' ', cleanTxt)
    
    return cleanTxt.strip()
```

```python
text = "RT @user: Check this link http://example.com 😎🔥 #Python #AI!!!"
print(cleanResume(text))
```

```python
df['Resume']= df['Resume'].apply(lambda x: cleanResume(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')

tfidf.fit(df['Resume'])
requiredText = tfidf.transform(df['Resume'])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(requiredText,df['Category'],test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score
clf = OneVsRestClassifier(KNeighborsClassifier())
clf.fit(x_train,y_train)
ypred = clf.predict(x_test)
print(accuracy_score(y_test,ypred))
```
